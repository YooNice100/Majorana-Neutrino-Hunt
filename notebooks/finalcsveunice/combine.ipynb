```python
import os
import pandas as pd

FEATURE_DIR = "."  # already inside finalcsveunice

print("Current directory:", os.getcwd())
print("Files found:", os.listdir(FEATURE_DIR))

feature_files = [
    "ED_train_all.csv",
    "HWP_train_all.csv",
    "LQ80_train_all.csv",
    "PPR_train_all.csv",
    "SC_train_all.csv"
]

# load first feature file
df = pd.read_csv(feature_files[0])
print("Loaded:", feature_files[0], df.shape)

# merge remaining features
for fname in feature_files[1:]:
    temp = pd.read_csv(fname)
    print("Merging:", fname, temp.shape)
    df = df.merge(temp, on="id", how="inner")

print("\nFinal merged shape:", df.shape)
print(df.head())

# save combined dataset
output_path = "train_all_features.csv"
df.to_csv(output_path, index=False)

print("\nSaved combined training CSV to:")
print(os.path.abspath(output_path))
```
