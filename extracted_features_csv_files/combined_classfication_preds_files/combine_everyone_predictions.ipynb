```python
import pandas as pd
```

```python
low_avse_df = pd.read_csv('predicted_low_avse_test.csv')
high_avse_df = pd.read_csv('high_avse_test_preds.csv')
dcr_avse_df = pd.read_csv('psd_label_dcr_predictions_nn.csv')
lq_df = pd.read_csv('psd_label_lq_predictions.csv')
```

```python
combined_df = low_avse_df.merge(high_avse_df, on='id').merge(dcr_avse_df, on='id').merge(lq_df, on='id')

combined_df = combined_df.rename(columns={
    "predicted_label_low_avse": "pred_low_avse",
    "high_avse_preds": "pred_high_avse",
    "predicted_label_dcr": "pred_dcr",
    "predicted_label_lq": "pred_lq",
})

combined_df
```

```python
combined_df[
    (combined_df["pred_low_avse"] == 1) &
    (combined_df["pred_high_avse"] == 1) &
    (combined_df["pred_dcr"] == 1) &
    (combined_df["pred_lq"] == 1)
]
```

```python
combined_df.to_csv("combined_classification_predictions.csv", index=False)
```
