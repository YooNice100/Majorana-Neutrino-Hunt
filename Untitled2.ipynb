```python
import h5py, os, numpy as np
import matplotlib.pyplot as plt

train_files = ["MJD_Train_0.hdf5", "MJD_Train_1.hdf5", "MJD_Train_12.hdf5"]

def list_keys(filename, label="train"):
    with h5py.File(filename, "r") as f:
        first_key = list(f.keys())[0]
        print(f"\n=== {label.upper()} FILE: {filename} ===")
        print("Number of rows:", f[first_key].shape[0])
        print("Columns (datasets):")
        for k in f.keys():
            print("  ", k)

def show_random_event(filename):
    with h5py.File(filename, "r") as f:
        W   = np.array(f["raw_waveform"])
        E   = np.array(f["energy_label"])
        Llo = np.array(f["psd_label_low_avse"])
        Lhi = np.array(f["psd_label_high_avse"])
        Ld  = np.array(f["psd_label_dcr"])
        Llq = np.array(f["psd_label_lq"])
        tp0 = np.array(f["tp0"])
        det = np.array(f["detector"])
        run = np.array(f["run_number"])
        ids = np.array(f["id"])

        idx = np.random.randint(0, W.shape[0])
        w   = W[idx]
        t   = np.arange(len(w))  

        plt.figure(figsize=(7,3))
        plt.plot(t, w)
        plt.title(f"{os.path.basename(filename)}  random waveform (id={ids[idx]})")
        plt.xlabel("Sample index")
        plt.ylabel("ADC")
        plt.tight_layout()
        plt.show()

        print("Data Labels:")
        print(f"  Energy:               {E[idx]}")
        print(f"  psd_label_low_avse:   {Llo[idx]}")
        print(f"  psd_label_high_avse:  {Lhi[idx]}")
        print(f"  psd_label_dcr:        {Ld[idx]}")
        print(f"  psd_label_lq:         {Llq[idx]}")
        print(f"  tp0:                  {tp0[idx]}")
        print(f"  detector:             {det[idx]}")
        print(f"  run_number:           {run[idx]}")
        print(f"  id:                   {ids[idx]}")

for f in train_files:
    if not os.path.exists(f):
        print(f"[!] File not found in this folder: {f}")
        continue
    list_keys(f, "train")
    show_random_event(f)
```

```python
import numpy as np

def compute_avse(raw_waveform, energy_label, n_baseline=50):
    N = raw_waveform.shape[0]
    AvsE = np.full(N, np.nan, float)
    A    = np.full(N, np.nan, float)

    for i in range(N):
        w = raw_waveform[i].astype(float)
        E = float(energy_label[i])
        b = np.mean(w[:n_baseline])
        w0 = w - b

        cur = np.diff(w0)  # per-sample slope
        Ai = np.max(cur) if cur.size else np.nan

        A[i] = Ai
        AvsE[i] = (Ai / E) if E > 0 else np.nan

    return AvsE, A
```

```python
import numpy as np, h5py, glob
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
from sklearn.metrics import roc_auc_score, roc_curve

def load_many_hdf5(fnames):
    arrays = {}
    for fp in fnames:
        with h5py.File(fp, "r") as f:
            for k in f.keys():
                arr = np.array(f[k])
                if k not in arrays:
                    arrays[k] = [arr]
                else:
                    arrays[k].append(arr)
    return {k: np.concatenate(v, axis=0) for k, v in arrays.items()}

wanted = ["MJD_Train_0.hdf5", "MJD_Train_1.hdf5", "MJD_Train_12.hdf5"]
existing = [f for f in wanted if len(glob.glob(f)) > 0]
if not existing:
    existing = sorted(glob.glob("MJD_Train_*.hdf5"))[:1]  # at least one

data = load_many_hdf5(existing)

raw_waveform        = data["raw_waveform"]       # (N, T)
energy_label        = data["energy_label"]       # (N,)
psd_label_low_avse  = data["psd_label_low_avse"] # (N,)
psd_label_dcr       = data["psd_label_dcr"]      # (N,)
psd_label_lq        = data["psd_label_lq"]       # (N,)

print("Loaded events:", raw_waveform.shape[0], "  samples per waveform:", raw_waveform.shape[1])

# 1) Simple AvsE feature 
def compute_avse_simple(raw_waveform, energy_label, n_baseline=50):
    """
    AvsE using peak-to-baseline amplitude.
    Returns:
      AvsE  (N,)
      A     (N,) amplitude (peak - baseline)
    """
    base = np.mean(raw_waveform[:, :n_baseline], axis=1)
    peak = np.max(raw_waveform, axis=1)
    amp  = peak - base
    E    = energy_label.astype(float)
    E[E <= 0] = np.nan
    AvsE = amp / E
    return AvsE, amp

AvsE_simple, A_simple = compute_avse_simple(raw_waveform, energy_label)

# 2) Pick a label to compare groups
label = psd_label_low_avse  

# 3) Hist + MannWhitney + ROC 
m  = np.isfinite(AvsE_simple)
x1 = AvsE_simple[(label==1) & m]
x0 = AvsE_simple[(label==0) & m]

plt.figure()
bins = np.linspace(np.nanpercentile(AvsE_simple, 1), np.nanpercentile(AvsE_simple, 99), 60)
plt.hist(x1, bins=bins, alpha=0.55, label='label=1')
plt.hist(x0, bins=bins, alpha=0.55, label='label=0')
plt.title("AvsE  Histogram"); plt.xlabel("AvsE"); plt.ylabel("count"); plt.legend(); plt.tight_layout()

if (len(x1) > 0) and (len(x0) > 0):
    u, p = mannwhitneyu(x1, x0, alternative='two-sided')
    print(f"MannWhitney on AvsE: U={u:.0f}, p={p:.3e}, n1={len(x1)}, n0={len(x0)}")

    y = label[m].astype(int)
    f = AvsE_simple[m]
    if len(np.unique(y)) == 2:
        auc = roc_auc_score(y, f)
        fpr, tpr, _ = roc_curve(y, f)
        print(f"AvsE AUC = {auc:.3f}")
        plt.figure(); plt.plot(fpr, tpr); plt.plot([0,1],[0,1],'--')
        plt.xlabel('FPR'); plt.ylabel('TPR'); plt.title(f'ROC  AvsE (AUC={auc:.3f})'); plt.tight_layout()
```

```python
import numpy as np
from scipy.signal import find_peaks, peak_widths, savgol_filter

def gradient_features(raw_waveform, n_baseline=50, sg_win=7, sg_poly=2):
    
    N = raw_waveform.shape[0]
    area1  = np.zeros(N, float)
    area2  = np.zeros(N, float)
    ratio  = np.zeros(N, float)
    width1 = np.zeros(N, float)

    for i in range(N):
        w  = raw_waveform[i].astype(float)
        b  = np.mean(w[:n_baseline])
        w0 = w - b

        x = savgol_filter(w0, sg_win, sg_poly, mode="interp") if len(w0) >= sg_win else w0
        d = np.diff(x)  

        peaks, _ = find_peaks(d, prominence=np.max(d)*0.05 if np.max(d)>0 else 0.0)
        if peaks.size == 0:
            area1[i]=area2[i]=ratio[i]=width1[i]=0.0
            continue

        if peaks.size == 1:
            wres = peak_widths(d, peaks, rel_height=0.5)
            w_main = wres[0][0] if len(wres[0]) else 0.0
            width1[i] = float(w_main)
            area_main = float(np.trapezoid(np.clip(d,0,None)))  # crude total positive area
            area1[i]  = area_main
            area2[i]  = 0.0
            ratio[i]  = area1[i] / (area2[i] + 1e-9)
        else:
            lobes = []
            for p in peaks:
                L = p
                while L>0 and d[L] > 0: L -= 1
                R = p
                while R < d.size-1 and d[R] > 0: R += 1
                lobe = np.clip(d[L:R+1], 0, None)
                lobes.append((p, float(np.trapezoid(lobe)), L, R))

            lobes.sort(key=lambda t: t[1], reverse=True)
            p1, a1, L1, R1 = lobes[0]
            area1[i]  = a1
            width1[i] = float(R1 - L1 + 1)

            a2 = lobes[1][1] if len(lobes) > 1 else 0.0
            area2[i] = a2
            ratio[i] = area1[i] / (area2[i] + 1e-9)

    return area1, area2, ratio, width1
```

```python
from scipy.stats import mannwhitneyu
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

# compute features
GradAreaMain, GradArea2, GradAreaRatio, GradWidthMain = gradient_features(raw_waveform)

# pick a ground-truth-ish label to compare (e.g., psd_label_low_avse as SSE-like=1)
y = psd_label_low_avse.astype(int)

def show_hist_and_roc(feat, name):
    mask = np.isfinite(feat)
    f = feat[mask]; yv = y[mask]
    # hist
    plt.figure()
    bins = np.linspace(np.nanpercentile(f,1), np.nanpercentile(f,99), 80)
    plt.hist(f[yv==1], bins=bins, alpha=0.6, label='label=1')
    plt.hist(f[yv==0], bins=bins, alpha=0.6, label='label=0')
    plt.title(f'{name}: hist by label'); plt.legend(); plt.tight_layout(); plt.show()

    # MW + AUC
    try:
        U,p = mannwhitneyu(f[yv==1], f[yv==0], alternative='two-sided')
    except ValueError:
        U,p = np.nan, np.nan
    auc = roc_auc_score(yv, f) if len(np.unique(yv))==2 else np.nan
    print(f'{name}: MannWhitney U={U:.0f}, p={p:.3e}, AUC={auc:.3f}')

show_hist_and_roc(GradAreaRatio, "GradAreaRatio")
show_hist_and_roc(GradWidthMain, "GradWidthMain")
```

```python
def compute_hfer(waveform, frac_high=0.25, n_baseline=50):
    w = np.asarray(waveform, dtype=float)
    if w.size == 0:
        return np.nan

    # baseline & normalize
    b = np.mean(w[:min(n_baseline, len(w))])
    w0 = w - b
    peak = np.max(w0)
    if not np.isfinite(peak) or peak <= 0:
        return np.nan
    x = w0 / peak

    # FFT magnitude
    mag = np.abs(np.fft.rfft(x))
    if mag.size <= 1:
        return np.nan  # nothing beyond DC
    mag_no_dc = mag[1:]  # drop DC

    # take the highest-frequency tail (top fraction)
    K = mag_no_dc.size
    k0 = int(max(0, np.floor((1.0 - frac_high) * K)))
    num = np.sum(mag_no_dc[k0:])
    den = np.sum(mag_no_dc)
    return float(num / den) if den > 0 else np.nan
```
